```python
import numpy as np
import matplotlib.pyplot as plt

# Sample image (for demonstration, let's create a synthetic edge image)
image = np.zeros((100, 100))
image[:, 50:] = 255

# Compute the gradient using finite difference
gradient_x = np.diff(image, axis=1)  # Gradient along x-axis
gradient_y = np.diff(image, axis=0)  # Gradient along y-axis

# Pad the result to match original image size
gradient_x = np.pad(gradient_x, ((0, 0), (0, 1)), mode='constant')
gradient_y = np.pad(gradient_y, ((0, 1), (0, 0)), mode='constant')

# Plot the gradients
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.title('Gradient along x-axis')
plt.imshow(gradient_x, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Gradient along y-axis')
plt.imshow(gradient_y, cmap='gray')

plt.show()
```
